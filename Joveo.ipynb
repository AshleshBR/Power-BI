```python
def update_contractual_margin(row):
    if row['Contractual Margin Type']==1:
        return row['Contractual Margin Information']
    if row['Contractual Margin Type']==2:
        return row['Contractual Margin Information']
    if row['Contractual Margin Type']==3:
        if row['TV']<=75000:
            return 5
        elif row['TV']>=150000:
            return 10
        else:
            return 7
    elif row['Contractual Margin Type']==4:
        if row['TV']<=20000:
            return 7.5
        elif row['TV']>=70000:
            return 4
        else:
            return 6
    elif row['Contractual Margin Type']==5:
        if row['TV']<=5000:
            return 5
        else:
            return 1500

data['Contractual Margin Type2'] = data.apply(update_contractual_margin, axis=1).astype(float)
```

```python
def update_contractual_margin2(row):
    if row['Contractual Margin Type']==1:
        return row['TV']*row['Contractual Margin Type2']/100
    if row['Contractual Margin Type']==2:
        return row['Contractual Margin Type2']
    if row['Contractual Margin Type']==3:
            return row['TV']*row['Contractual Margin Type2']/100
    elif row['Contractual Margin Type']==4:
        if row['TV']<=20000:
            return row['TV']*7.5/100
        elif row['TV']>=70000:
            return (20000*(7.5/100))+(50000*(6/100))+((row['TV']-70000)*4/100)
        else:
            return (20000*(7.5/100))+((row['TV']-20000)*6/100)
    elif row['Contractual Margin Type']==5:
        if row['TV']<=5000:
            return row['TV']*5/100
        else:
            return row['Contractual Margin Type2']

data['Contractual Margin'] = data.apply(update_contractual_margin2, axis=1).astype(float)
```

```python
data['GR Forecast Delta %']=((data['GR']-data['GR Forecast'])/data['GR Forecast'])*100
```

```python
data['NR(Net Revenue']=data['Publisher Commission']+data['Trading Margin']+data['Contractual Margin']
```
